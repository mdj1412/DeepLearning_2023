## 리스트로부터 배열 생성
NumPy 배열은 파이썬 리스트와 비슷하지만 전체 원소에 대해 계산을 수행할 때 
빠릅니다.

* 이유 : list는 random access가 빠르다. ( 각 type이 다르기 때문에 한 개씩 사용해야한다. )
* list : 추가와 삭제가 편하다.
* numpy : 각 type이 같다.

```python
# 키와 몸무게 리스트
height = [1.87,  1.87, 1.82, 1.91, 1.90, 1.85]
weight = [81.65, 97.52, 95.25, 92.98, 86.18, 88.45]

# NumPy 사용시 필요한 것
import numpy as np

# 리스트를 NumPy 배열로 변환하는 방법
np_height = np.array(height)
np_weight = np.array(weight)
```

```python
# 키 리스트의 내용과 타입 확인
print(height)
print(type(height))
```

## 원소별 계산<br>
체질량지수 (BMI) 계산을 해봅시다. (체중 / 키^2)<br>
18.5 이하 저체중, 18.5 ~ 25 정상, 25 이상 과체중

```python
# BMI 계산하여 리스트에 저장하기
bmi = [] # BMI를 저장할 빈 리스트 생성

for i in range(len(height)):
  bmi.append(weight[i] / height[i] ** 2)

print(bmi)
print(type(bmi))
```

```python
# BMI 계산하여 NumPy 배열에 저장하기
np_height = np.array(height)
np_weight = np.array(weight)

# Calculate bmi
np_bmi = np_weight / np_height ** 2

# Print the result
print(np_bmi)
print(type(np_bmi))
```

BMI 계산 시 위의 두 방식 중 어떤 것이 편한가요? 대용량 데이터 처리 시 속도는 NumPy가 확실히 빠릅니다.

다음으로 리스트와 NumPy의 덧셈을 비교해봅시다.

```python
# 리스트 덧셈
print(weight + height)

# NumPy 배열 덧셈
print(np_weight + np_height)
```
